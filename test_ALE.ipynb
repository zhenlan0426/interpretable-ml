```python
import numpy as np
import shap
import matplotlib.pyplot as plt
import seaborn 
from local_ICE import *
```

```python
def model(X):
    return X[:,0] * X[:,1]
```

```python
X = np.random.randn(1000,2)
y = model(X)
X_corr = np.cumsum(X,1)
y_corr = model(X_corr)
```

independent X

```python
shap.partial_dependence_plot(0, model, X)
```

```python
ale_np(X,model,[0],include_CI=False)
```

```python
local_ICE(model,X,0)
```

```python
# M-plot
seaborn.regplot(x=X[:,0],y=y,scatter=False,lowess=True)
```

Dependent X

```python
ale_np(X_corr,model,[0],include_CI=False)
```

```python
shap.partial_dependence_plot(0, model, X_corr)
```

```python
local_ICE(model,X_corr,0)
```

```python
# M-plot
seaborn.regplot(x=X_corr[:,0],y=y_corr,scatter=False,lowess=True)
```
