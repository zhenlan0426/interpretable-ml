```python
import shap
import sklearn
from local_ICE import local_ICE
```

```python
# a classic housing price dataset
X,y = shap.datasets.boston()
X100 = shap.utils.sample(X, 100) # 100 instances for use as the background distribution


import xgboost
model_xgb = xgboost.XGBRegressor(n_estimators=400, max_depth=6).fit(X, y)

# explain the GAM model with SHAP
explainer_xgb = shap.Explainer(model_xgb, X100)
shap_values_xgb = explainer_xgb(X)


i = 4 
shap.plots.scatter(shap_values_xgb[:,i])
shap.partial_dependence_plot(i, model_xgb.predict, X)
local_ICE(model_xgb.predict,X.values,i)
```
